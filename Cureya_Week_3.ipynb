```python
X = pd.DataFrame(np.c_[df1['LSTAT'], df1['RM']], columns=['LSTAT', 'RM'])
Y = df1['MEDV']
```

##**Splitting the Data into Training & Testing**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=5)
```

**Prediction of Price using following Algorithms:**


> a. Linear Regression


> b. KNN Algorithm





#**a. Linear Regression**

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
# Model Evaluation for TRAINING SET:
print('Model Performance for Training Set:')
y_pred = model.predict(X_train)
# Root Mean Square Error 
rmse = (np.sqrt(mean_squared_error(Y_train, y_pred)))
print('RMSE is {}'.format(rmse))
# r-squared score 
r2 = r2_score(Y_train, y_pred)
print('R2 Score is {}'.format(r2))
```

```python
# Model Evaluation for TESTING SET:
print('Model Performance for Testing Set:')
y_pred = model.predict(X_test)
# Root Mean Square Error of the the model
rmse = (np.sqrt(mean_squared_error(Y_test, y_pred)))
print('RMSE is {}'.format(rmse))
# r-squared score of the model
r2 = r2_score(Y_test, y_pred)
print('R2 Score is {}'.format(r2))
```

##**Price Prediction using Linear Regression**

```python
plt.scatter(Y_test, y_pred)
plt.title('Original vs Predicted Prices', fontsize = 22)
plt.xlabel('Original Prices')
plt.ylabel('Predicted Prices')
plt.show()
```

#**b. Using KNN Algorithm**

```python
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
```

```python
boston = load_boston()
print('Shape of Boston Dataset:{}'.format(boston.data.shape))
```

```python
print(boston.keys())
```

```python
X_train, X_test, y_train, y_test = train_test_split(boston.data, boston.target)
```

```python
bos = pd.DataFrame(boston['data'])
bos.columns = boston['feature_names']
bos['Price'] = boston['target']
bos.head()
```

```python
 def get_r_squared(max_neighbors=10):
    train_score = []
    test_score = []
    models = []
    for neighbors in range(1, max_neighbors+1):
        model = KNeighborsRegressor(n_neighbors=neighbors)
        model.fit(X_train, y_train)
        train_score.append(model.score(X_train, y_train))
        test_score.append(model.score(X_test, y_test))
        models.append(model)
    return train_score, test_score, models
```

```python
 train_score, test_score, models = get_r_squared()
neighbors = range(1, 11)
plt.plot(neighbors, train_score, label="Training $r^2$")
plt.plot(neighbors, test_score, label="Testing $r^2$")
plt.xlabel("Neighbors")
plt.ylabel("$r^2$")
plt.title("KNN Boston Housing Prices")
plt.legend()
```

```python
print("Training r2 for 2 neigbors: {:.2f}".format(train_score[1]))
print("Testing r2 for 2 neighbors: {:.2f}".format(test_score[1]))
#assert max(test_score) == test_score[1]
```
